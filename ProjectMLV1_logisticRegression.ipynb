```python
data  = pd.read_csv('./singleDataSet.csv')
sport = data.article[data.classe == 'sport'].sum()
economie = data.article[data.classe == 'economie'].sum()
politique = data.article[data.classe == 'politique'].sum()
culture = data.article[data.classe == 'culture'].sum()
articles = [sport,politique,culture,economie]
concatData = pd.DataFrame(data = {'category':['sport','politique','culture','economie'],'document':articles})
vect = CountVectorizer()
T = vect.fit_transform(concatData.document)
T = T.toarray()
T[T>1] = 1
N = T.sum(axis = 0)>2
wordsToRemove = np.array(vect.get_feature_names_out())[N]
```

```python
data.head(20)
```

```python
data_train,data_test = train_test_split(data,train_size = 0.7,random_state = 1)
dtr = netoyage(np.array(data_train.article),wordsToRemove)
dts = netoyage(np.array(data_test.article),wordsToRemove)
train_cleanArticles = pd.DataFrame(data={'article':dtr})
test_cleanArticles = pd.DataFrame(data={'article':dts})
train_cleanArticles.index = data_train.index
test_cleanArticles.index = data_test.index
data_train.article = train_cleanArticles
data_test.article = test_cleanArticles
```

```python
# parseur = CountVectorizer(binary=True)# avec stopWrods et binary
# parseur = CountVectorizer()
parseur = TfidfVectorizer()
X_train = parseur.fit_transform(data_train.article)
mdtTrain = X_train.toarray()
X_test = parseur.transform(data_test.article)
mdtTest = X_test.toarray()
mdtTest
parseur.vocabulary_
```

```python
clf = LogisticRegression()
# clf = KNeighborsClassifier(n_neighbors=5)
# clf = RandomForestClassifier(n_estimators=200,random_state=1)
# clf = RandomForestClassifier(n_estimators=200,max_depth=10)
clf.fit(mdtTrain,data_train.classe)
score1 = clf.score(mdtTrain,data_train.classe)
score2 = clf.score(mdtTest,data_test.classe)
print(f'score sur train {score1.round(2)}')
print(f'score sur train {score2.round(2)}\n')
```
